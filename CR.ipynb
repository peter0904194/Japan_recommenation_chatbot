```python
from selenium import webdriver
from selenium.webdriver.edge.service import Service
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from bs4 import BeautifulSoup
import time

# Edge WebDriver 경로 설정
service = Service('C:/V/msedgedriver.exe')
driver = webdriver.Edge(service=service)

# 대상 URL로 이동
url = "https://m.place.naver.com/my/5e353a338f87a842bc2e4924/reviewfeed?sort=VISIT_DATE_TIME_DESC&mediaFilter=WITH_MEDIA&reviewId=64cf3344beae8c7940113506&v=1"
driver.get(url)

# 스크롤하여 리뷰를 순차적으로 출력
scroll_pause_time = 2  # 스크롤 사이 대기 시간
last_height = driver.execute_script("return document.body.scrollHeight")
review_count = 0  # 이전에 출력한 리뷰 개수 추적

while True:
    # 현재 페이지 HTML 가져오기
    page_source = driver.page_source
    soup = BeautifulSoup(page_source, 'html.parser')

    # 현재 페이지의 리뷰 블록 가져오기
    review_blocks = soup.find_all("div", class_="_898uc8")

    # 새로운 리뷰만 출력
    for i, review_block in enumerate(review_blocks[review_count:], start=review_count + 1):
        # 1. 매장 이름 추출
        store_name = review_block.find("span", class_="pui__pv1E2a")
        print(f"[리뷰 {i}] 매장 이름:", store_name.get_text() if store_name else "정보 없음")

        # 2. 저장 여부 정보 추출
        saved_status = review_block.find("span", class_="pui__WUm6H8")
        print(f"[리뷰 {i}] 저장 여부:", saved_status.get_text() if saved_status else "정보 없음")

        # 3. 리뷰 내용 추출
        review = review_block.find("a", class_="pui__xtsQN-")
        print(f"[리뷰 {i}] 리뷰 내용:", review.get_text() if review else "리뷰 정보 없음")

        # 4. 대기 관련 정보 추출
        waiting_info = [span.get_text() for span in review_block.find_all("span", class_="pui__jhpEyP")]
        print(f"[리뷰 {i}] 추가 정보:", ", ".join(waiting_info) if waiting_info else "추가 정보 없음")

        # 5. 방문일 정보 추출
        visit_date_elem = review_block.find("time", attrs={"aria-hidden": "true"})
        visit_date = visit_date_elem.text if visit_date_elem else "방문일 정보 없음"
        print(f"[리뷰 {i}] 방문일:", visit_date)

        print("\n" + "="*50 + "\n")

    # 새로운 리뷰를 출력한 후, 출력된 리뷰 수를 갱신
    review_count = len(review_blocks)

    # 페이지 끝까지 스크롤
    driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")

    # 스크롤 후 페이지 로딩 대기
    time.sleep(scroll_pause_time)

    # 새로운 페이지 높이 확인
    new_height = driver.execute_script("return document.body.scrollHeight")
    if new_height == last_height:
        # 더 이상 새로운 내용이 없으면 종료
        break
    last_height = new_height

# WebDriver 종료
driver.quit()
```

```python
#################################################### 찐 코드 #############################################3





from selenium import webdriver
from selenium.webdriver.edge.service import Service
from bs4 import BeautifulSoup
import pandas as pd
import time

# Edge WebDriver 경로 설정
service = Service('C:/V/msedgedriver.exe')
driver = webdriver.Edge(service=service)

# URL 리스트 생성
#위에서 함

# 데이터를 저장할 빈 리스트 생성
data = []

# 각 URL에 대해 순차적으로 접근하여 리뷰 크롤링
for idx, url in enumerate(urls, start=1):
    # 대상 URL로 이동
    driver.get(url)
    time.sleep(2)  # 페이지 로딩 대기
    
    # 스크롤하여 리뷰를 순차적으로 출력
    scroll_pause_time = 2  # 스크롤 사이 대기 시간
    last_height = driver.execute_script("return document.body.scrollHeight")
    review_count = 0  # 이전에 출력한 리뷰 개수 추적
    
    while True:
        # 현재 페이지 HTML 가져오기
        page_source = driver.page_source
        soup = BeautifulSoup(page_source, 'html.parser')
        
        # 현재 페이지의 리뷰 블록 가져오기
        review_blocks = soup.find_all("div", class_="_898uc8")
        
        # 새로운 리뷰만 저장
        for i, review_block in enumerate(review_blocks[review_count:], start=review_count + 1):
            # 리뷰 정보를 데이터 리스트에 저장
            store_name = review_block.find("span", class_="pui__pv1E2a")
            saved_status = review_block.find("span", class_="pui__WUm6H8")
            review = review_block.find("a", class_="pui__xtsQN-")
            waiting_info = [span.get_text() for span in review_block.find_all("span", class_="pui__V8F9nN")]
            visit_date_elem = review_block.find("time", attrs={"aria-hidden": "true"})
            visit_date = visit_date_elem.text if visit_date_elem else "방문일 정보 없음"
            review2 = [span.get_text() for span in review_block.find_all("span", class_="pui__jhpEyP")]
            

            # 유저명 생성 (user1, user2, ...)
            user_name = f"user{idx}"

            # 데이터 저장
            data.append({
                "user": user_name,
                "name": store_name.get_text() if store_name else "정보 없음",
                "category": saved_status.get_text() if saved_status else "정보 없음",
                "visit_date": visit_date,
                "info": ", ".join(waiting_info) if waiting_info else "추가 정보 없음",
                "more_info_": ", ".join(review2) if review2 else "추가 정보 없음",
                "review": review.get_text() if review else "리뷰 정보 없음"
            })
        
        # 새로운 리뷰를 출력한 후, 출력된 리뷰 수를 갱신
        review_count = len(review_blocks)
        
        # 페이지 끝까지 스크롤
        driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
        
        # 스크롤 후 페이지 로딩 대기
        time.sleep(scroll_pause_time)
        
        # 새로운 페이지 높이 확인
        new_height = driver.execute_script("return document.body.scrollHeight")
        if new_height == last_height:
            # 더 이상 새로운 내용이 없으면 종료
            break
        last_height = new_height

# WebDriver 종료
driver.quit()

# DataFrame 생성 및 CSV로 저장
df = pd.DataFrame(data)
df.to_csv("reviews.csv", index=False, encoding="utf-8-sig")
print("리뷰 데이터를 reviews.csv 파일로 저장했습니다.")
```
