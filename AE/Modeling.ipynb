Multi Autoencoder 모델

```python
# 사용자-아이템 상호작용 매트릭스를 구성하기 위한 전처리
# 'category0'와 'category2'를 결합한 새로운 열 생성
data['item'] = data['category0'] + '_' + data['category2']

# 사용자와 아이템 데이터를 기반으로 상호작용 매트릭스 생성
interaction_matrix = data.pivot_table(
    index='user',    # 사용자
    columns='item',  # 아이템 (카테고리 조합)
    aggfunc='size',  # 카운트(상호작용 횟수)
    fill_value=0     # NaN은 0으로 채움
)
interaction_matrix
```

데이터 분리

```python
import numpy as np
from scipy.sparse import csr_matrix
from sklearn.model_selection import train_test_split

# 희소 행렬 형태로 변환 (Multi-VAE에 적합)
interaction_sparse = csr_matrix(interaction_matrix.values)

# 학습 및 검증 세트로 분리
train_data, test_data = train_test_split(
    interaction_sparse, test_size=0.2, random_state=42
)

# 데이터 요약
train_data_shape = train_data.shape
test_data_shape = test_data.shape

train_data_shape, test_data_shape
```

모델 정의

```python
import torch
from torch import nn, optim
from torch.utils.data import DataLoader, TensorDataset
from scipy.sparse import csr_matrix

#모델 정의
class MultiAutoencoder(nn.Module):
    def __init__(self, p_dims):
        super(MultiAutoencoder, self).__init__()
        self.encoder = nn.Sequential(
            nn.Linear(p_dims[2], p_dims[1]),
            nn.ReLU(),
            nn.Linear(p_dims[1], p_dims[0])
        )
        self.decoder = nn.Sequential(
            nn.Linear(p_dims[0], p_dims[1]),
            nn.ReLU(),
            nn.Linear(p_dims[1], p_dims[2]),
            nn.Sigmoid()
        )

    def forward(self, x):
        z = self.encoder(x)
        return self.decoder(z)

# 하이퍼파라미터 설정
n_items = train_data.shape[1]
p_dims = [200, 600, n_items]
device = torch.device("cpu")  # CPU 사용

# 데이터 정규화: 0과 1 사이로 변환
def normalize_data(sparse_data):
    dense_data = sparse_data.toarray().astype(np.float32)
    dense_data = np.clip(dense_data, 0, 1)
    return dense_data

# DataLoader 생성
def create_dataloader(sparse_data, batch_size=64):
    dense_data = sparse_data.toarray().astype(np.float32)
    dataset = TensorDataset(torch.tensor(dense_data))
    return DataLoader(dataset, batch_size=batch_size, shuffle=True)

# 데이터 로드 및 전처리
train_loader = create_dataloader(csr_matrix(normalize_data(train_data)))
test_loader = create_dataloader(csr_matrix(normalize_data(test_data)))

# 모델 초기화
model = MultiAutoencoder(p_dims).to(device)
criterion = nn.BCELoss(reduction="sum")
optimizer = optim.Adam(model.parameters(), lr=0.001)

# 학습 루프
n_epochs = 100
for epoch in range(n_epochs):
    model.train()
    train_loss = 0
    for batch in train_loader:
        inputs = batch[0].to(device)
        optimizer.zero_grad()
        outputs = model(inputs)
        loss = criterion(outputs, inputs)
        loss.backward()
        optimizer.step()
        train_loss += loss.item()

    train_loss /= len(train_loader.dataset)
    print(f"Epoch {epoch + 1}/{n_epochs}, Loss: {train_loss:.4f}")

print("학습 완료!")
```

Test Data -> 카테고리 TOP10 

```python
# 테스트 데이터 확인
print("Test data shape:", test_data.shape)
print("Sample test data (dense):", normalize_data(test_data)[:5])

# 추천 결과 생성 디버깅
def generate_recommendations_debug(model, data_loader, n_recommendations=10):
    model.eval()
    recommendations = {}
    print("Starting recommendation generation...")
    with torch.no_grad():
        for i, batch in enumerate(data_loader):
            inputs = batch[0].to(device)
            outputs = model(inputs).cpu().numpy()  # 모델 예측 값
            for idx, user_scores in enumerate(outputs):
                user_id = i * data_loader.batch_size + idx  # 사용자 고유 ID
                seen_items = inputs[idx].cpu().numpy() > 0
                user_scores[seen_items] = -1  # 이미 본 아이템의 점수를 낮게 설정
                top_items = np.argsort(user_scores)[-n_recommendations:][::-1]
                if len(top_items) > 0:
                    recommendations[int(user_id)] = list(top_items)
                else:
                    print(f"No recommendations for user_id {user_id}")
    print("Finished recommendation generation.")
    return recommendations

# 추천 결과 생성 및 디버깅
recommendations_debug = generate_recommendations_debug(model, test_loader, n_recommendations=10)

# 추천 결과를 데이터프레임으로 변환 및 확인
if recommendations_debug:  # 추천 결과가 비어있지 않은 경우만 처리
    recommendation_df_debug = pd.DataFrame.from_dict(recommendations_debug, orient='index', columns=[f"Item_{i+1}" for i in range(10)])
    print(recommendation_df_debug.head())
else:
    print("No recommendations generated. Please check the test data and model.")
```

성능 평가

Precision@K : 추천된 K개의 항목 중에서 사용자가 실제로 좋아한 항목의 비율, 추천 품질이 얼마나 높은지를 평가
</br>
보통: 0.10~0.20
좋음: 0.20~0.30
우수: 0.30 이상
</br>
Reca?ll@K : 사용자가 실제로 좋아한 모든 항목 중에서 추천 리스트에 포함된 비율, 사용자가 좋아할만한 항목을 얼마나 잘 포괄했는지 평가
</br>
보통: 0.05~0.10
좋음: 0.10~0.20
우수: 0.20 이상
</br>
NDCG@K (Normalized Discounted Cumulative Gain) : 추천된 항목의 순위 품질을 평가, 용자가 좋아할 항목이 상위에 올수록 추천 품질이 좋음을 나타냄
</br
보통: 0.10~0.20
좋음: 0.20~0.40
우수: 0.40 이상

```python
# Precision@K, Recall@K, NDCG 계산 함수
def evaluate_recommendations(recommendations, test_data, k=10):
    precision, recall, ndcg = 0.0, 0.0, 0.0
    num_users = len(recommendations)

    for user_id, recommended_items in recommendations.items():
        # 실제 상호작용 아이템
        true_items = np.where(test_data[user_id].toarray()[0] > 0)[0]

        # 교집합: 추천된 아이템과 실제 아이템의 교집합
        hits = np.intersect1d(recommended_items[:k], true_items)

        # Precision@K
        precision += len(hits) / k

        # Recall@K
        if len(true_items) > 0:
            recall += len(hits) / len(true_items)

        # NDCG
        dcg = 0.0
        for rank, item in enumerate(recommended_items[:k]):
            if item in true_items:
                dcg += 1 / np.log2(rank + 2)
        ideal_dcg = sum(1 / np.log2(rank + 2) for rank in range(min(len(true_items), k)))
        ndcg += dcg / ideal_dcg if ideal_dcg > 0 else 0.0

    # 평균 계산
    precision /= num_users
    recall /= num_users
    ndcg /= num_users

    return precision, recall, ndcg

# 추천 결과 평가
precision, recall, ndcg = evaluate_recommendations(recommendations_debug, test_data, k=10)

# 평가 결과 출력
print(f"Precision@10: {precision:.4f}")
print(f"Recall@10: {recall:.4f}")
print(f"NDCG@10: {ndcg:.4f}")
```

예: Precision@10 = 0.15라면, 추천한 10개 중 1.5개 정도가 사용자가 선호한 항목임
예: Recall@10 = 0.09라면, 사용자가 좋아한 항목 중 9%만 추천 리스트에 포함됨
예: NDCG@10 = 0.18이라면, 상위 10개의 추천 항목 중 순위가 중요한 영향을 미쳤음을 보여줌

```python
# 추천 결과를 다시 생성
recommendations_debug = generate_recommendations_debug(model, test_loader, n_recommendations=10)

# 사용자별 리뷰 개수 계산
user_review_counts = np.array(test_data.sum(axis=1)).flatten()

# 사용자 그룹화: 리뷰 개수에 따라 그룹 나누기
high_review_users = np.where(user_review_counts >= 10)[0]  # 리뷰 10개 이상
low_review_users = np.where((user_review_counts > 0) & (user_review_counts < 10))[0]  # 리뷰 1~9개
very_high_review_users = np.where(user_review_counts >= 20)[0]  # 리뷰 20개 이상

# 추천 결과 필터링: 각 그룹에 해당하는 사용자 추천만 추출
def filter_recommendations_by_group(recommendations, group_users):
    return {user_id: recommendations[user_id] for user_id in group_users if user_id in recommendations}

# 각 그룹에 대한 추천 결과
high_review_recommendations = filter_recommendations_by_group(recommendations_debug, high_review_users)
low_review_recommendations = filter_recommendations_by_group(recommendations_debug, low_review_users)
very_high_review_recommendations = filter_recommendations_by_group(recommendations_debug, very_high_review_users)

# 성능 평가 함수 호출
precision_high, recall_high, ndcg_high = evaluate_recommendations(high_review_recommendations, test_data, k=10)

precision_very_high, recall_very_high, ndcg_very_high = evaluate_recommendations(very_high_review_recommendations, test_data, k=10)

# 결과 출력
print(f"High Review Users (10+ Reviews):")
print(f"  Precision@10: {precision_high:.4f}, Recall@10: {recall_high:.4f}, NDCG@10: {ndcg_high:.4f}")

print(f"Very High Review Users (20+ Reviews):")
print(f"  Precision@10: {precision_very_high:.4f}, Recall@10: {recall_very_high:.4f}, NDCG@10: {ndcg_very_high:.4f}")
```

모델을 이용해 모든 유저에게 카테고리 추천

```python
# 모델을 평가 모드로 전환
model.eval()

# 전체 유저에 대해 1개의 추천 아이템을 생성
def recommend_for_all_users(model, interaction_matrix):
    recommendations = []

    with torch.no_grad():
        for user_idx in range(interaction_matrix.shape[0]):
            # 해당 유저의 상호작용 데이터 가져오기
            user_data = torch.tensor(interaction_matrix.iloc[user_idx].values).float().to(device)

            # 예측 수행
            output = model(user_data.unsqueeze(0))  # 모델에 입력으로 넣기 위해 차원 추가

            # 가장 높은 예측값을 가진 아이템을 추천
            _, top_item_idx = torch.max(output, dim=1)

            # 추천된 아이템의 인덱스 저장
            recommendations.append((user_idx + 1, top_item_idx.item()))  # 1-based indexing

    return recommendations

# 추천 생성
recommendations = recommend_for_all_users(model, interaction_matrix)

# 추천 결과 출력 (유저 ID와 추천된 아이템의 인덱스)
for user, item in recommendations:
    print(f"User {user} -> Recommended item: {interaction_matrix.columns[item]}")
```

각 유저에게 추가적인 정보 (가장 많이 언급된 것들)

```python
import pandas as pd
from collections import Counter

# Function to get the top 3 most frequent words/phrases from '상황/인물' and 'more_info_' for each user
def get_top_words_per_user(data):
    top_words_per_user = {}

    for user_id in data['user'].unique():
        # Filter data for the user
        user_data = data[data['user'] == user_id]

        # Get words/phrases for '상황/인물' and 'more_info_'
        situation_words = user_data['상황/인물'].dropna().str.split(',').sum()
        more_info_words = user_data['more_info_'].dropna().str.split(',').sum()

        # Get the frequency count of words/phrases
        situation_word_counts = Counter(situation_words)
        more_info_word_counts = Counter(more_info_words)

        # Get top 3 words/phrases for both columns
        top_situation_words = situation_word_counts.most_common(3)
        top_more_info_words = more_info_word_counts.most_common(4)

        # Store results in a dictionary
        top_words_per_user[user_id] = {
            "상황/인물_top": top_situation_words,
            "more_info_top": top_more_info_words
        }

    return top_words_per_user

# Get top words/phrases per user
top_words_per_user = get_top_words_per_user(data)

# Remove '음식이 맛있어요' from the results
def remove_word_from_top(top_words_per_user, word_to_remove):
    top_words_per_user_filtered = {}

    for user_id, words in top_words_per_user.items():
        # Filter out the unwanted word/phrase
        filtered_situation_words = [word for word in words['상황/인물_top'] if word[0] != word_to_remove]
        filtered_more_info_words = [word for word in words['more_info_top'] if word[0] != word_to_remove]

        top_words_per_user_filtered[user_id] = {
            "상황/인물_top": filtered_situation_words,
            "more_info_top": filtered_more_info_words
        }

    return top_words_per_user_filtered

# Remove '음식이 맛있어요' from the top words
top_words_per_user_filtered = remove_word_from_top(top_words_per_user, '음식이 맛있어요')

# Convert to DataFrame for better visualization
top_words_df_filtered = pd.DataFrame.from_dict(top_words_per_user_filtered, orient='index')

top_words_df_filtered
```

```python
# 추천된 음식과 상위 3개 상황/인물, more_info_ 결합
def merge_recommendations_with_top_words(recommendations, top_words_df):
    merged_data = []

    for user, item in recommendations:
        # 추천된 음식 이름 가져오기
        recommended_item = interaction_matrix.columns[item]

        # 해당 유저의 top words 가져오기
        top_situation_words = top_words_df.loc[user, '상황/인물_top']
        top_more_info_words = top_words_df.loc[user, 'more_info_top']

        # 결과 결합
        merged_data.append({
            'user': user,
            'recommended_item': recommended_item,
            '상황/인물_top': top_situation_words,
            'more_info_top': top_more_info_words
        })

    return pd.DataFrame(merged_data)

# 추천 음식과 상위 단어 결합
merged_df = merge_recommendations_with_top_words(recommendations, top_words_df_filtered)
merged_df
```
